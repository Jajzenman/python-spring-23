```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
webpage = requests.get("https://www.nypl.org/research/research-catalog/bib/b11343493?originalUrl=https%3A%2F%2Fcatalog.nypl.org%2Frecord%3Db11343493~S1")
source = webpage.content
soup = BeautifulSoup(source, 'lxml')
```

```python
for item in soup.find_all('script'):
    print(item.text)
```
